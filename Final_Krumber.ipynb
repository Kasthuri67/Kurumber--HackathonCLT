```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
df_health = pd.read_csv('qol-health.csv')
df_health.info()
```

```
df_health.head()
```

```
df_health.isnull().sum()
```

##Data Cleaning

```
df_health.duplicated().sum()
```

```
df_health.dropna(inplace=True)
```

```
df_health.isnull().sum()
```

```
df_health.info()
```

```
#df_health.replace(' ', regex=True)
df_health.columns = map(str.lower, df_health.columns)
```

```
df_health.describe()
```

```
df_npa = pd.read_csv('qol-npa-mapping.csv')
df_npa.head(2)
```

```
df_npa.drop(columns='Briarcreek-Woodland', inplace=True)
```

```
df_npa.head(2)
```

```
df_combined = df_health.merge(df_npa, left_on='npa', right_on = 'npa', how='inner')
df_combined.head()
```

```
df_combined.info()
```

```
interested_cols = ['npa','name','low_birthweight_2012','low_birthweight_2014','births_to_adolescents_2012','births_to_adolescents_2014','prenatal_care_2012','prenatal_care_2014']
health_new_df = df_combined[interested_cols]
health_new_df.head(2)
```

```
health_new_df.rename(index=str, columns={'name': 'npa_name'})
health_new_df.head(2)
```

```
health_new_df.describe()
```

## Birth to Adolescents vs Low Birthweight

```
# plotting graph
ax = sns.regplot(x=health_new_df['births_to_adolescents_2012'],y=health_new_df['low_birthweight_2012'])

#setup the title and the labels of the scatter plot.
ax.set_title("Correlation between Low Birthweight and Birth to Adolescents - 2012",fontsize=13)
ax.set_xlabel("Birth to Adolescents",fontsize=12)
ax.set_ylabel("Low Birthweight",fontsize=12)

# setup the figure size.
sns.set(rc={'figure.figsize':(13,6)})
# set style
sns.set_style("whitegrid")

#find the correlation between them.
data_corr = health_new_df.corr()
print("Correlation between Low Birthweight and Birth to Adolescents - 2012 : ",data_corr.loc['births_to_adolescents_2012','low_birthweight_2012'])
```

```
# plotting graph
ax = sns.regplot(x=health_new_df['births_to_adolescents_2014'],y=health_new_df['low_birthweight_2014'])

#setup the title and the labels of the scatter plot.
ax.set_title("Correlation between Low Birthweight and Birth to Adolescents - 2014",fontsize=13)
ax.set_xlabel("Birth to Adolescents",fontsize=12)
ax.set_ylabel("Low Birthweight",fontsize=12)

# setup the figure size.
sns.set(rc={'figure.figsize':(13,6)})
# set style
sns.set_style("whitegrid")

#find the correlation between them.
data_corr = health_new_df.corr()
print("Correlation between Low Birthweight and Birth to Adolescents - 2014 : ",data_corr.loc['births_to_adolescents_2014','low_birthweight_2014'])
```
